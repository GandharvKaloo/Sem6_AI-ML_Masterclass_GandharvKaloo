```python
from sklearn.preprocessing import LabelEncoder as LE
catg_cols_label = ['Location', 'Model']
label_encoders = {}
```

```python
for col in catg_cols_label:
    le = LE()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le
```

```python
from sklearn.preprocessing import OneHotEncoder as OHE
catg_cols_ohe = ["Fuel_Type", "Transmission", "Owner_Type"]
one_hot_encoders = {}

for col in catg_cols_ohe:
    ohe = OHE(drop='first', sparse_output=False)  # New parameter name
    reshaped_col = X[col].values.reshape(-1, 1)  # Reshaping into 2D array
    encoded_col = ohe.fit_transform(reshaped_col)
    
    # Convert encoded array into DataFrame
    encoded_df = pd.DataFrame(encoded_col, columns=ohe.get_feature_names_out([col]))
    
    # Reset index to match DataFrame
    encoded_df.index = X.index
    
    # Concatenate with original DataFrame
    X = pd.concat([X, encoded_df], axis=1)
    
    # Save encoder
    one_hot_encoders[col] = ohe

# Drop original categorical columns
X = X.drop(columns=catg_cols_ohe)
```
