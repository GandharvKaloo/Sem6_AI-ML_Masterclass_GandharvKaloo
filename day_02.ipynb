### MATPLOT - VIZUALIZATIONS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#creating line plot
x = np.arange(0, 10, 0.1)
y = 2*x + 5

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = 2*x + 5

plt.plot(x,y)
plt.show()

fig = plt.figure(figsize = (20,5))

plt.plot(x,y, linewidth = 2.0, linestyle = ":", color = 'r', alpha = 0.5, marker = "o")
plt.title("line plot customization")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["line1"])
plt.grid(True)
plt.show()
```

### Sub Plots

```python
x = np.arange(0,10,0.1)
y1 = 2*x + 5
y2 = 3*x + 10
y3 = 4*x + 15
y4 = 5*x + 20



plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('Graph1')

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('Graph2')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('Graph3')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('Graph4')
plt.show()
```

### Bar Plot

```python
data = {"apples": 20, "mangoes": 10, "lemon": 50, "oranges":58}

names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10,5), facecolor = 'lightblue')

plt.bar(names, values, color = ['red', 'orange'])
plt.show()
```

### Scatter Plots
