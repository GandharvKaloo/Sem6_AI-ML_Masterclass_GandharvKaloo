```python
data = pd.read_csv("used_cars_maruti.csv")
```

```python
x_features = ["Location", "Fuel_Type", "Transmission", "Owner_Type", "Seats", "Age", "Model", "Mileage", "Power", "KM_Driven"]
```

```python
cat_var = ["Location",
"Fuel_Type",
"Transmission",
"Owner_Type",
"Model"]
```

```python
num_var = list(set(x_features) - set(cat_var))
num_var
```

```python
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
from sklearn.model_selection import train_test_split
X=data[x_features]
y=data['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
ohe_encoder = OneHotEncoder(handle_unknown='ignore')
cat_transformer = Pipeline(steps=[('oheencoder', ohe_encoder)])
```

```python
cat_transformer
minmax_scaler=MinMaxScaler()
num_transformer= Pipeline(steps=[('scaler', minmax_scaler)])
```

```python
preprocessor = ColumnTransformer(transformers = [('numerical', num_transformer,num_var),('categorical',cat_transformer,cat_var)])
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=5,weights='uniform')
```

```python
knn_pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('regression', knn)])
knn_pipeline.fit(X_train, y_train)
```

```python
y_pred = knn_pipeline.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error , r2_score
np.round(r2_score(y_test,y_pred), 5)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(knn_pipeline, X_train, y_train, cv=5, scoring='r2')
```
