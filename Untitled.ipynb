```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled_data = scaler.fit_transform(data[['calories', 'sodium', 'alcohol', 'cost']])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(4, random_state=100 ,n_init='auto')
```

```python
kmeans.fit(X_scaled_data)
```

```python
kmeans.labels_
```

```python
data['cluster'] = kmeans.labels_
```
